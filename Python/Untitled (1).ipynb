```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path=r"C:\Users\Jeremy\Desktop"
os.chdir(path)
nfl_data = pd.read_csv("nflsalary.csv")
nfl_data
nfl_data = nfl_data.loc[:, ~nfl_data.columns.str.contains('^Unnamed')]
nfl_data

pay_wins = nfl_data[['Games Won', 'Active']]
pay_wins = pay_wins.astype({"Games Won":"int","Active":"int"})
pay_wins
```

```python
matrix = np.array(pay_wins.values,'float')
X = matrix[:,0]
y = matrix[:,1]
print(matrix)
```

```python
plt.plot(X,y,'bo')
plt.xlabel("Games Won")
plt.ylabel("Payroll")
```

```python
def best_fit(X, y):

    xbar = sum(X)/len(X)
    ybar = sum(y)/len(y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b

# solution
a, b = best_fit(X, y)
plt.plot(X,y,'bo')
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit)

print("For 0 wins, total team salary will be: $"+str(np.round(a))+". For each additional win, teams will have to spend $"+str(np.round(b))+".")
```
